```python
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer
import numpy as np
from scipy.special import softmax
import csv
import urllib.request
```

```python
task1='sentiment'
MODEL1 = f"cardiffnlp/twitter-roberta-base-{task1}"
tokenizer1 = AutoTokenizer.from_pretrained(MODEL1)

task2='emotion'
MODEL2 = f"cardiffnlp/twitter-roberta-base-{task2}"
tokenizer2 = AutoTokenizer.from_pretrained(MODEL2)

task3='irony'
MODEL3 = f"cardiffnlp/twitter-roberta-base-{task3}"
tokenizer3 = AutoTokenizer.from_pretrained(MODEL3)
```

```python
# download label mapping sentiment
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/{task1}/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels1 = [row[1] for row in csvreader if len(row) > 1]
labels1
```

```python
# download label mapping emotion
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/{task2}/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels2 = [row[1] for row in csvreader if len(row) > 1]
labels2
```

```python
# download label mapping irony
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/{task3}/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels3 = [row[1] for row in csvreader if len(row) > 1]
labels3
```

```python
model1 = TFAutoModelForSequenceClassification.from_pretrained(MODEL1)
model2 = TFAutoModelForSequenceClassification.from_pretrained(MODEL2)
model3 = TFAutoModelForSequenceClassification.from_pretrained(MODEL3)
```

```python
# this function return labels for sentiment
def return_labels1(text):
    encoded_input = tokenizer1(text, return_tensors='tf')
    output = model1(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    label1 = labels1[ranking[0]]
    return label1
```

```python
# this function return labels for emotions
def return_labels2(text):
    encoded_input = tokenizer2(text, return_tensors='tf')
    output = model2(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    label2 = labels2[ranking[0]]
    return label2 
```

```python
# this function return labels for irony
def return_labels3(text):
    encoded_input = tokenizer3(text, return_tensors='tf')
    output = model3(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    label3 = labels3[ranking[0]]
    return label3 
```

```python
# this function return probabilities to analize how "sure" the model is
def return_prob1(text):
    encoded_input = tokenizer1(text, return_tensors='tf')
    output = model1(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    return np.sort(scores)[-1]
```

```python
# this function return probabilities to analize how "sure" the model is
def return_prob2(text):
    encoded_input = tokenizer2(text, return_tensors='tf')
    output = model2(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    return np.sort(scores)[-1]
```

```python
# this function return probabilities to analize how "sure" the model is
def return_prob3(text):
    encoded_input = tokenizer3(text, return_tensors='tf')
    output = model3(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    return np.sort(scores)[-1]
```

```python
detokenized_df['label_sentiment'] = detokenized_df['pre_processed'].apply(lambda x: return_labels1(x))
detokenized_df['probability_sentiment'] = detokenized_df['pre_processed'].apply(lambda x: return_prob1(x))

detokenized_df['label_emotion'] = detokenized_df['pre_processed'].apply(lambda x: return_labels2(x))
detokenized_df['probability_emotion'] = detokenized_df['pre_processed'].apply(lambda x: return_prob2(x))

detokenized_df['label_irony'] = detokenized_df['pre_processed'].apply(lambda x: return_labels3(x))
detokenized_df['probability_irony'] = detokenized_df['pre_processed'].apply(lambda x: return_prob3(x))

detokenized_df
```

```python
official_df = pd.concat([depression_df[['renderedContent']], detokenized_df], axis = 1)
official_df
```
